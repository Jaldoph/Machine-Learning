```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.model_selection import RandomizedSearchCV
import os
print(os.listdir("/home/cspooner/equake-LANL/input/"))
```

```python
train = pd.read_csv("/home/cspooner/equake-LANL/input/train.csv", dtype={'acoustic_data': np.int16, 'time_to_failure': np.float32})
```

```python
rows = 2_055_000
segments = int(np.floor(train.shape[0] / rows))

X_train = pd.DataFrame(index=range(segments), dtype=np.float32,
                       columns=['ave', 'std', 'max', 'min','sum','absave', 'ptp', 'pct', 'gradientmin',
                                'gradientmean','gradientstd', 'gradientmax', 'gradientpct', 
                                'absstd','absmax', 'absmin','abssum','absgradient', 'abspct' ])
y_train = pd.DataFrame(index=range(segments), dtype=np.float32,
                       columns=['time_to_failure'])
abstrain= train.abs()

for segment in range(segments):
    seg = train.iloc[segment*rows:segment*rows+rows]
    x = (seg['acoustic_data'].values)
    y = seg['time_to_failure'].values.mean()
    
    y_train.loc[segment, 'time_to_failure'] = y
    
    X_train.loc[segment, 'ptp'] = x.ptp()
    X_train.loc[segment, 'ave'] = x.mean()
    X_train.loc[segment, 'std'] = x.std()
    X_train.loc[segment, 'max'] = x.max()
    X_train.loc[segment, 'min'] = x.min()
    X_train.loc[segment, 'sum'] = x.sum()
    X_train.loc[segment, 'pct'] = x.var()
    
    X_train.loc[segment, 'gradientmean'] = np.gradient(x).mean()
    X_train.loc[segment, 'gradientstd'] = np.gradient(x).std()
    X_train.loc[segment, 'gradientmin'] = np.gradient(x).min()
    X_train.loc[segment, 'gradientmax'] = np.gradient(x).max()
    X_train.loc[segment, 'gradientpct'] = np.gradient(x).var()
    
    
for segment in range(segments):
    seg = abstrain.iloc[segment*rows:segment*rows+rows]
    x = (seg['acoustic_data'].values)
    y = seg['time_to_failure'].values.mean()
    
    y_train.loc[segment, 'time_to_failure'] = y
    
    
    X_train.loc[segment, 'abspct'] = x.var()
    X_train.loc[segment, 'absave'] = x.mean()
    X_train.loc[segment, 'absstd'] = x.std()
    X_train.loc[segment, 'absmax'] = x.max()
    X_train.loc[segment, 'absmin'] = x.min()
    X_train.loc[segment, 'abssum'] = x.sum()
    X_train.loc[segment, 'absgradient'] = np.gradient(x).mean()
    #X_train.loc[segment, 'avg'] = x.median()
    #X_train.loc[segment, 'prod'] = x.prod()

```
